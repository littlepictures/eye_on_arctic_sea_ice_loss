```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('vis.csv')
df['extent'] = df['extent'] * -1
df.head()
```

```python
chart = alt.Chart(df).mark_area(
    color="white",
    interpolate='linear',
    line=False,
    opacity=1
).encode(
    x=alt.X("date:T", axis=None),
    y=alt.Y("extent:Q", axis=None)
).properties(
    width=600,
    height=600
).configure(
    background='#0b2031'
).configure_view(
    strokeWidth=0
)

chart
```

```python
## Bauhaus Style Poster

# main chart
chart = alt.Chart(df).mark_area(
    color="#F6F6F6",
    interpolate='linear',
    line=False,
    opacity=1
).encode(
    x=alt.X("date:T", axis=None),
    y=alt.Y("extent:Q", axis=None)
).properties(
    width=763,
    height=900
)

# title text
title = alt.Chart().mark_text(
    text='arctic sea ice decline',
    size=35,
    font='Helvetica',
    fontWeight='bold',
    color='#fff',
    align='left'
).encode(
    x=alt.value(20),
    y=alt.value(900)
)

# logo text
logo = alt.Chart().mark_text(
    text='treva labs',
    size=20,
    font='Helvetica',
    fontWeight='bold',
    color='#fff',
    align='right'
).encode(
    x=alt.value(740),
    y=alt.value(890)
)

# detail text
detail = alt.Chart().mark_text(
    text='1980/2020',
    size=20,
    font='Helvetica',
    fontWeight='lighter',
    color='#fff',
    align='left'
).encode(
    x=alt.value(20),
    y=alt.value(950)
)

# Layering shapes and text
poster = alt.layer(chart, title, detail + logo).configure_view(
    strokeWidth=0,
    width=763,
    height=1080
).configure(
    padding={"left": 0, "top": 0, "right": 0, "bottom": 100},
    background='#0b2031'
).configure_view(
    strokeWidth=0
)

poster
```
